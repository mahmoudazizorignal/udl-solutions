**Problem 4.8∗** Figure $4.9$ shows the activations in the three hidden units of a shallow network (as in figure $3.3$). The slopes in the hidden units are $1.0$, $1.0$, and $-1.0$, respectively, and the “joints” in the hidden units are at positions $1/6$, $2/6$, and $4/6$. Find values of $ϕ_0$, $ϕ_1$, $ϕ_2$, and $ϕ_3$ that will combine the hidden unit activations as $ϕ_0 + ϕ_1h_1 + ϕ_2h_2 + ϕ_3h_3$ to create a function with four linear regions that oscillate between output values of zero and one. The slope of the leftmost region should be positive, the next one negative, and so on. How many linear regions will we create if we compose this network with itself? How many will we create if we compose it with itself $K$ times?

<center>
    <img src="assets/4.9.jpeg">

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# define our parameters
theta_10, theta_11, theta_20, theta_21, theta_30, theta_31 = -1/6, 1, -2/6, 1, 4/6, -1
phi_0, phi_1, phi_2, phi_3 = 0.7, -1.5, 1, -1

# define relu activation
def relu(z):
    return np.maximum(z, 0)

# define some input
x = np.arange(0.0, 1.0, 0.001)

# calculte the network
h1 = relu(theta_10 + theta_11 * x)
h2 = relu(theta_20 + theta_21 * x)
h3 = relu(theta_30 + theta_31 * x)
y = phi_0 + phi_1 * h1 + phi_2 * h2 + phi_3 * h3

# plot the units
fig = plt.figure(figsize=(8,6))
gs = gridspec.GridSpec(2, 3, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax1.set_ylim(-1.0, 1.0)
ax1.set_title(r"$h_1 = a[\theta_{10} + \theta_{11}x]$")
ax1.plot(x, h1, color="red")

ax2 = fig.add_subplot(gs[0, 1])
ax2.set_ylim(-1.0, 1.0)
ax2.set_title(r"$h_2 = a[\theta_{20} + \theta_{21}x]$")
ax2.plot(x, h2, color="green")

ax3 = fig.add_subplot(gs[0, 2])
ax3.set_ylim(-1.0, 1.0)
ax3.set_title(r"$h_3 = a[\theta_{30} + \theta_{31}x]$")
ax3.plot(x, h3, color="blue")

ax4 = fig.add_subplot(gs[1, :])
ax4.set_ylim(0, 1.0)
ax4.set_title(r"$y = \phi_0 + \phi_1 h_1 + \phi_2 h_2 + \phi_3 h_3$")
ax4.plot(x, y, color="cyan")

plt.tight_layout()
plt.show()
```

If we compose this shallow network on itself, it will generate $4 \times 4$ linear region

Composing the network with itself $K$ times will yield $4^K$ linear region

2) The output layer will have: $\ohm_K \in \mathbb{R}^{1 \times D}$, and $\beta_K \in \mathbb{R}$, that is a total of $1\cdot D + 1.$
3) The remaining $(K - 1)$ hidden layer, each one will have: $\ohm_k \in \mathbb{R}^{D \times D}$, and $\beta_k \in \mathbb{R}^{D}$, for $k \in [1, K - 1[$, and that is a total of $(K - 1)\cdot D \cdot D + (K - 1)\cdot D.$
